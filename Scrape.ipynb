```
provinces = ['ha-noi', 'ho-chi-minh', 'da-nang', 'ba-ria-vung-tau', 'lam-dong', 'khanh-hoa',
            'thua-thien-hue', 'ninh-binh', 'binh-duong', 'dong-nai', 'quang-nam', 'quang-ninh',
            'hai-phong', 'can-tho', 'binh-thuan', 'hai-duong', 'bac-ninh', 'lao-cai',
            'nam-dinh', 'vinh-phuc', 'ben-tre', 'long-an', 'hoa-binh', 'ha-nam',
            'tien-giang', 'hung-yen', 'phu-tho', 'thai-binh', 'tay-ninh', 'kien-giang',
            'thai-nguyen', 'ninh-thuan', 'binh-dinh', 'quang-ngai', 'bac-giang', 'lang-son',
            'quang-tri', 'phu-yen', 'vinh-long', 'dak-lak', 'an-giang', 'dong-thap',
            'binh-phuoc', 'son-la', 'soc-trang', 'ca-mau', 'tuyen-quang', 'ha-giang',
            'bac-lieu', 'gia-lai', 'hau-giang', 'tra-vinh', 'dak-nong', 'kon-tum',
            'cao-bang', 'thanh-hoa', 'lai-chau', 'bac-kan', 'nghe-an', 'quang-binh',
            'ha-tinh', 'yen-bai', 'dien-bien']
len(provinces)
```

```
# Scrape data from the website
gody = pd.DataFrame(columns = ['Province', 'Tourists', 'Destination', 'Rate', 'Visitors'])
for name in provinces[60:]:
    link_href = 'https://gody.vn/chau-a/viet-nam/' + name + '/diem-den'
    driver.get(link_href)
    time.sleep(20)

    # Extract province
    province = driver.find_element_by_xpath('//*[@id="app-root"]/div[2]/div[1]/h1').text

    # Extract tourists
    tourists = driver.find_element_by_xpath('//*[@id="app-root"]/div[2]/div[2]/div[2]/span/span').text
    
    info = driver.find_element_by_xpath('//*[@class="w-fit d-block col-md-12 list-destination p-0 m-0"]').text
    list_info = info.split('\nĐánh giá ngay\n')
    for destination_info in list_info:
        splited = destination_info.split('\n')
        
        # Extract destination
        destination = splited[0]

        # Extract rate
        rate = splited[1]

        # Extract visitor
        visitors = splited[2].split(' ')[0]

        gody.loc[len(gody)] = [province, tourists, destination, rate, visitors]
```

```
gody.to_csv('./gody 60-end.csv', index=False)
```
